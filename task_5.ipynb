### 1)Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.Then create a horizontal bar chart showing total runs scored in their debut 5 years.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2000, 2002, 2004, 2006, 2008]
sachin_runs = [1200, 1300, 1250, 1400, 1350]
kohli_runs  = [0,    0,    500,  900,  1200]
sehwag_runs = [700,  1100, 1300, 1000, 800]

x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(9,6))
plt.bar(x - width,sachin_runs,width=width,color='orange',label='Sachin')
plt.bar(x,kohli_runs,width=width,color='royalblue',label='Kohli')
plt.bar(x + width,sehwag_runs,width=width,color='mediumseagreen',label='Sehwag')

for i in range(len(x)):
    plt.text(x[i] - width, sachin_runs[i] + 20, str(sachin_runs[i]), ha='center')
    plt.text(x[i], kohli_runs[i] + 20, str(kohli_runs[i]), ha='center')
    plt.text(x[i] + width, sehwag_runs[i] + 20, str(sehwag_runs[i]), ha='center')

plt.title("Year-wise Runs Comparison of Sachin, Kohli, and Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")

plt.xticks(x,years)
plt.legend()
plt.grid(axis='y',alpha=0.2)
plt.tight_layout()

plt.show()
```

```python
players = ['Sachin', 'Kohli', 'Sehwag']
total_runs_5yrs = [4500, 4300, 4000]

plt.figure(figsize=(8,6))
plt.barh(players,total_runs_5yrs,color=['#87CEEB','#98FB98','#F08080'])

for i in range(len(total_runs_5yrs)):
    plt.text(total_runs_5yrs[i] + 50, i, str(total_runs_5yrs[i]),va='center')

plt.title("Total Runs in First 5 Years of Debut")
plt.xlabel("Total Runs")
plt.ylabel("Players")

plt.grid(axis='x',alpha=0.2)
plt.tight_layout()

plt.show()
```

### 2) Create a pie chart showing the market share of mobile OS (Android, iOS, others).Then recreate the same using a horizontal bar chart and observe which is easier to understand.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['Xiaomi', 'Samsung', 'Vivo', 'Realme', 'Oppo', 'Others']
market_share = [22, 18, 17, 15, 10, 10]

plt.style.use('ggplot')
colors = ['#4CAF50', '#2196F3', '#FFC107', '#9C27B0', '#FF5722', '#607D8B']

explodes= [0.1,0,0,0,0,0]
plt.pie(market_share,labels=labels,colors=colors,autopct='%1.1f%%',explode=explodes,shadow=True,startangle=90)
plt.title("Android Brand Market Share in India (2025)")
plt.show()
```

```python
labels = ['Xiaomi', 'Samsung', 'Vivo', 'Realme', 'Oppo', 'Others']
market_share = [22, 18, 17, 15, 10, 10]

colors =['#FF5733','#1F77B4','#9C27B0','#2ECC71','#FF9800','#BDBDBD']

plt.style.use('default')
plt.barh(labels,market_share,color=colors)

for i in range(len(market_share)):
    plt.text(market_share[i] + 1, i, f'{market_share[i]}%', va='center')
    
plt.title("Android Brand Market Share in India (2025)")
plt.xlabel("Market Share (%)")
plt.ylabel("Brands")
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()

plt.show()
```
